1.2 3D coordinate frame

$$     	p  = \begin{bmatrix} p_x \\ p_y \\ p_z \end{bmatrix} $$

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#  
p = (3, 5, 7)

# 3D  
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

#  
ax.scatter(p[0], p[1], p[2], color='red', s=100)  #  
ax.text(p[0]+0.5, p[1], p[2], f'p{p}', fontsize=12)

#  
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

#    (   )
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_zlim(0, 10)

# 
ax.set_title('3D coordinate')

plt.show()
```
