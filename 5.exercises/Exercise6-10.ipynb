### 6. 
Prove that all length-preserving transforms in 2D are either rotations or rotations followed by a mirroring transform:$$ T(X)= \begin{bmatrix}-1 & 0  \\0 & 1 \\\end{bmatrix} X $$.

변환 $T(x) = A x$ 가 길이를 보존하면 모든 벡터 $x$에 대해:

$$
\| T(x) \| = \| x \| \quad \Rightarrow \quad \| A x \|^2 = \| x \|^2
$$

즉,

$$
x^T A^T A x = x^T x \quad \text{for all } x
$$

따라서,

$$
A^T A = I
$$

즉, $A$ 는 **직교 행렬** (orthogonal matrix)이다.

2D에서 직교 행렬 $A$는 항상 다음 두 가지 중 하나다:

- $\det(A) = +1$ → **회전 행렬**
- $\det(A) = -1$ → **회전 + 반사**

---

예: $T(x) = \begin{bmatrix} -1 & 0 \\\\ 0 & 1 \end{bmatrix} x$

이 행렬은 직교 행렬이고,

$$
A^T A = I, \quad \det(A) = -1
$$

이므로 **길이 보존**하지만 단순 회전은 아님. 이는 $x$축에 대해 **반사(mirroring)** 하는 변환이다.

---

결론: 2D에서 길이를 보존하는 모든 선형변환은

- 회전, 또는
- 회전 + 반사 (미러링)

중 하나이다.


### 7. 
Produce the transform (i.e., rotation matrix  R and translation  t) that rotates any point  x by  45∘ about the point  (1,0). What is the general form of a rotation of angle  θ about a center point  c?

주어진 중심점 $c = (1, 0)$에서 $45^\circ$ 회전

회전 행렬:

$$
R = \begin{bmatrix}
\cos\theta & -\sin\theta \\\\
\sin\theta & \cos\theta
\end{bmatrix}, \quad \theta = 45^\circ = \frac{\pi}{4}
$$

$$
R = \frac{1}{\sqrt{2}} \begin{bmatrix}
1 & -1 \\\\
1 & 1
\end{bmatrix}
$$

중심 $c$를 기준으로 회전하기 위해 평행이동 벡터 $t$는 다음과 같다:

$$
T(x) = R(x - c) + c = R x + (c - R c)
$$

따라서 전체 변환은:

$$
T(x) = R x + t, \quad t = c - R c
$$

여기서 $c = \begin{bmatrix} 1 \\\\ 0 \end{bmatrix}$ 이므로,

$$
R c = \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\\\ 1 \end{bmatrix}, \quad
t = \begin{bmatrix} 1 \\\\ 0 \end{bmatrix} - \frac{1}{\sqrt{2}} \begin{bmatrix} 1 \\\\ 1 \end{bmatrix}
= \begin{bmatrix} 1 - \frac{1}{\sqrt{2}} \\\\ -\frac{1}{\sqrt{2}} \end{bmatrix}
$$

---

일반적인 중심점 $c$에 대해 각도 $\theta$로 회전하는 강체변환은:

$$
T(x) = R(\theta) x + t, \quad \text{where } t = c - R(\theta) c
$$

또는

$$
T(x) = R(\theta)(x - c) + c
$$

이 형식은 회전 중심이 원점이 아닌 임의의 점일 때 사용된다.


### 8.
A mobile robot has a coordinate convention that  X
  is forward and  Y
  is to the left. It has an sweeping laser sensor attached to it that returns distance readings at all angles around the robot using a spinning mirror. The laser sensor gives perfect depth readings, and is attached so that the mirror's axis of rotation is centered at the point  (0.7,0.05)
  in the robot's  X,Y
  frame and oriented so that the  0∘
  reading is pointing in the robot's  X
  direction. At time  t
 , the laser detects an obstacle at 1 m distance at the  15∘
  reading. At time  t+1
 , the robot moves 0.5 m forward, 0.1 m to the left, and 20 ∘
  counter-clockwise. Assuming the obstacle does not move, at which angle will the sensor detect the obstacle, and at what distance reading?

장애물의 위치를 절대 좌표계에서 구한 뒤,  
로봇이 이동한 후의 센서 좌표계에서 해당 점의 각도와 거리를 다시 구한다.

1. 시점 $t$: 센서 기준 장애물 위치 (극좌표 → 직교좌표)

$$
p_s = \begin{bmatrix}
\cos 15^\circ \\\\
\sin 15^\circ
\end{bmatrix}
= \begin{bmatrix}
0.9659 \\\\
0.2588
\end{bmatrix}
$$

센서 위치 $c = (0.7, 0.05)$ 이므로, 로봇 기준 장애물 위치:

$$
p_r = c + p_s
= \begin{bmatrix} 0.7 \\\\ 0.05 \end{bmatrix} + \begin{bmatrix} 0.9659 \\\\ 0.2588 \end{bmatrix}
= \begin{bmatrix} 1.6659 \\\\ 0.3088 \end{bmatrix}
$$

2. 로봇 이동: $T_{\text{robot}}$

- 이동: $t = \begin{bmatrix} 0.5 \\\\ 0.1 \end{bmatrix}$
- 회전: $R = R(20^\circ)$

$$
R = \begin{bmatrix}
\cos 20^\circ & -\sin 20^\circ \\\\
\sin 20^\circ & \cos 20^\circ
\end{bmatrix}
= \begin{bmatrix}
0.9397 & -0.3420 \\\\
0.3420 & 0.9397
\end{bmatrix}
$$

새 로봇 기준에서 장애물 좌표:

$$
p_{\text{new robot}} = R^{-1} (p_r - t)
= R^T (p_r - t)
$$

계산:

$$
p_r - t = \begin{bmatrix} 1.6659 - 0.5 \\\\ 0.3088 - 0.1 \end{bmatrix}
= \begin{bmatrix} 1.1659 \\\\ 0.2088 \end{bmatrix}
$$

$$
R^T = \begin{bmatrix}
0.9397 & 0.3420 \\\\
-0.3420 & 0.9397
\end{bmatrix}
$$

$$
p' = R^T \cdot \begin{bmatrix} 1.1659 \\\\ 0.2088 \end{bmatrix}
= \begin{bmatrix}
0.9397 \cdot 1.1659 + 0.3420 \cdot 0.2088 \\\\
-0.3420 \cdot 1.1659 + 0.9397 \cdot 0.2088
\end{bmatrix}
= \begin{bmatrix}
1.1508 \\\\
-0.1916
\end{bmatrix}
$$

3. 센서 위치 보정: 센서는 여전히 로봇에서 $c$만큼 떨어져 있음

센서 좌표계에서 장애물:

$$
p_{\text{sensor}} = p' - c
= \begin{bmatrix} 1.1508 - 0.7 \\\\ -0.1916 - 0.05 \end{bmatrix}
= \begin{bmatrix} 0.4508 \\\\ -0.2416 \end{bmatrix}
$$

거리:

$$
d = \|p_{\text{sensor}}\| = \sqrt{0.4508^2 + (-0.2416)^2} \approx 0.511
$$

각도 (센서 기준에서 로봇의 X축 방향이 0도):

$$
\theta = \arctan2(-0.2416, 0.4508) \approx -28.2^\circ
$$

---

**결과:**

- 센서는 약 $-28.2^\circ$ 방향에서 장애물을 감지한다.
- 측정된 거리: 약 $0.511$ m


### 센서 기준에서 새롭게 측정되는 각도와 거리를 자동으로 계산하는 함수

```python
import numpy as np

def rotation_matrix(theta_deg):
    theta = np.radians(theta_deg)
    return np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta),  np.cos(theta)]
    ])

def transform_obstacle(sensor_pos, obs_r, obs_theta_deg, robot_move, robot_rotation_deg):
    theta_rad = np.radians(obs_theta_deg)
    obstacle_sensor = np.array([np.cos(theta_rad), np.sin(theta_rad)]) * obs_r
    obstacle_robot = sensor_pos + obstacle_sensor

    R = rotation_matrix(robot_rotation_deg)
    R_inv = R.T
    obstacle_new_robot = R_inv @ (obstacle_robot - robot_move)
    obstacle_sensor_new = obstacle_new_robot - sensor_pos

    distance = np.linalg.norm(obstacle_sensor_new)
    angle = np.degrees(np.arctan2(obstacle_sensor_new[1], obstacle_sensor_new[0]))

    return angle, distance

# --- 사용자 입력 ---
print("장애물 정보와 로봇 이동 정보를 입력하세요:")

obs_r = float(input("장애물까지의 거리 (m): "))
obs_theta_deg = float(input("장애물 각도 (°): "))
robot_dx = float(input("로봇의 전방 이동 거리 dx (m): "))
robot_dy = float(input("로봇의 좌측 이동 거리 dy (m): "))
robot_rotation_deg = float(input("로봇의 회전 각도 (°): (반시계 +, 시계 -) "))

# 센서 위치 고정
sensor_pos = np.array([0.7, 0.05])
robot_move = np.array([robot_dx, robot_dy])

# 계산
angle, distance = transform_obstacle(sensor_pos, obs_r, obs_theta_deg, robot_move, robot_rotation_deg)

# 출력
print(f"\n[결과] 센서 기준 장애물 위치:")
print(f"새로운 각도: {angle:.2f}°")
print(f"새로운 거리: {distance:.3f} m")
```
