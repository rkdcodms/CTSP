2.3 Rotaions in 3D



벡터 P를 축 V(단위벡터)를 중심으로 $\theta$ 만큼 돌리는 변환

$$ \bf p' = R \bf p $$


R : 3X3 회전 행렬, p: 원래 벡터 , p': 회전된 벡터

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.spatial.transform import Rotation as R

# 회전 행렬 생성 (예: Z축 기준 45도 회전 + Y축 기준 30도)
rot = R.from_euler('zy', [45, 30], degrees=True)  # Z-Y 축 순서
R_mat = rot.as_matrix()

# 기본 좌표축 (단위 벡터)
origin = np.array([0, 0, 0])
x_axis = np.array([1, 0, 0])
y_axis = np.array([0, 1, 0])
z_axis = np.array([0, 0, 1])

# 회전된 좌표축
x_rot = R_mat @ x_axis
y_rot = R_mat @ y_axis
z_rot = R_mat @ z_axis

# 3D 시각화
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.set_title("3D Rotation: Original Axes vs Rotated Axes")

# 기본 좌표축 (굵은 화살표)
ax.quiver(*origin, *x_axis, color='r', linewidth=3, label='X (Original)')
ax.quiver(*origin, *y_axis, color='g', linewidth=3, label='Y (Original)')
ax.quiver(*origin, *z_axis, color='b', linewidth=3, label='Z (Original)')

# 회전된 좌표축 (점선 또는 얇은 화살표)
ax.quiver(*origin, *x_rot, color='r', linestyle='dashed', linewidth=1, label="X'")
ax.quiver(*origin, *y_rot, color='g', linestyle='dashed', linewidth=1, label="Y'")
ax.quiver(*origin, *z_rot, color='b', linestyle='dashed', linewidth=1, label="Z'")

# 좌표축 설정
ax.set_xlim([-1.2, 1.2])
ax.set_ylim([-1.2, 1.2])
ax.set_zlim([-1.2, 1.2])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend(loc='upper left')

plt.tight_layout()
plt.show()
print("Rotated X axis (column 1):", x_rot)
print("Rotated Y axis (column 2):", y_rot)
print("Rotated Z axis (column 3):", z_rot)
```
