# 기초 기하 및 좌표 변환 용어 요약

---

### Point (점)  
공간의 한 위치를 나타내는 벡터  
예: $P = [1.0,\ 2.0,\ 3.0]$

---

### Displacement (변위)  
두 점 사이의 차이 (방향 + 거리)  
예: $q - p = [3, 0, -1]$

---

###  Direction (방향)  
정규화된 단위벡터 (방향만 있음)  
예: $[1, 0, 0]$ 또는 $[0.707, 0.707, 0]$

---

### Coordinates (좌표)  
어떤 좌표계 기준으로 표현된 수치 값  
예: $[3, 1, 2]$

---

### Coordinate Frame (좌표계)  
원점 + 직교하는 축(X, Y, Z)으로 구성된 기준  
모든 위치/방향은 이 기준으로 표현됨

---

### Linear Interpolation (선형 보간)  
점 $p$ 에서 $q$ 까지 일정한 속도로 이동

$$
y(u) = (1 - u)p + u q = p + u(q - p)
$$

- $u=0$일 때 $p$, $u=1$일 때 $q$

---

### Plane Equation (평면 방정식)  
모든 평면 위의 점 $x$ 는 다음 식을 만족:

$$
x \cdot n = c
$$

- $n$: 평면의 수직 벡터  
- $c$: 오프셋 (거리)

---

### Rotation (회전)  
거리를 보존하는 방향 변환  
$$
\| R x - R y \| = \| x - y \|
$$

---

### Rotation Matrix (회전 행렬)  
2×2 또는 3×3 행렬, 회전을 나타냄  
- 직교 행렬, $\det(R) = +1$

---

### SO(2), SO(3)  
- $\text{SO(2)}$: 2D 회전 집합  
- $\text{SO(3)}$: 3D 회전 집합

---

### Rigid Transform (강체 변환)  
회전 + 평행이동을 결합한 거리 보존 변환

---

### Homogeneous Coordinates (동차좌표)  
회전 + 이동을 행렬 곱 하나로 표현

$$
T = \begin{bmatrix}
R & t \\\\
0 & 1
\end{bmatrix}
$$

- 점: $\hat{p} = [x, y, z, 1]$  
- 방향: $\hat{v} = [v_x, v_y, v_z, 0]$
